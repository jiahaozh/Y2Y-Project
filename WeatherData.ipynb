```python
year = [2016,2017,2018]
url = []
for i in range(len(year)):
    if i != 2:
        for j in range(1,13):
            url.append("https://www.usclimatedata.com/climate/boston/massachusetts/united-states/usma0046/"+str(year[i])+'/'+str(j))
    else:
        for j in range(1,3):
            url.append("https://www.usclimatedata.com/climate/boston/massachusetts/united-states/usma0046/"+str(year[i])+'/'+str(j))
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

# wait = WebDriverWait(driver, 600)
driver = webdriver.Firefox()
badDays = []
goodDays = []
for i in range(len(url)):
    driver.get(url[i])
    # driver.implicitly_wait(10000) # seconds
    # xpath = '//*[@id="ui-id-4"]'
    # driver.find_element_by_xpath(xpath).click()
    # time.sleep(5)
    # xpath = '/html/body/div[1]/div/section/div[2]/div[5]/div[3]/div/table/tbody[2]/tr/td[2]/span'
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    table = soup.find_all('table')
    row_marker = 0
    # for row in table[5].find_all('tr'):
    # df = pd.DataFrame(columns = (0,6), index = range(0,30))
    myList = {}
    for row in table[5].find_all('tr'):
        column_marker = 0
        columns = row.find_all('td')
        innerlist = []
        for column in columns:
            innerlist.append(column.get_text())
    #         print(column.get_text())
        try:
            myList[innerlist[0]] = innerlist[1:5]
            if innerlist[2] < '32' or innerlist[3] > '1' or innerlist[4] > '3':
                badDays.append(innerlist)
            else:
                goodDays.append(innerlist)
        except:
            pass
```
