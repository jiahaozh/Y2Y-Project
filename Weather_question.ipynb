```python
# creat a dictionary to define : which day ,who come
dic_date = {}
for row in lottery_entry.itertuples():
    if type(row[2]) == float:
        break
    else:
        date = str(row[8].strftime('%y-%m-%d'))
        if date not in dic_date:
            dic_date[date] = [row[1]]
        else:
            dic_date[date].append(row[1])
```

```python
dic_date
```

```python
# cluster their status based on the long-term plan
flag = []
for row in plan.itertuples():   
    longTermPlan = row[9]
    if type(longTermPlan) != float:
        if 'No plans' in longTermPlan:
            flag.append(0)
        elif 'Shelter' in longTermPlan:
            flag.append(0)
        else:
            flag.append(1)
    else:
        flag.append(0)
plan['flag'] = flag
```

```python
DateFlag = {}
for row in plan.itertuples():
    if type(row[2]) == str:
        date = str(row[3].strftime('%y-%m-%d'))
        if date not in DateFlag:
            DateFlag[date] = [[row[2],row[11]]]
        else:
            DateFlag[date].append([row[2],row[11]])
    
```

```python
dic_people = {}
for row in lottery_entry.itertuples():
    if type(row[2]) == float:
        break
    else:
        person = str(row[1])
        if person not in dic_people:
            dic_people[person] = [str(row[8].strftime('%y-%m-%d'))]
        else:
            dic_people[person].append(str(row[8].strftime('%y-%m-%d')))
```

```python
from datetime import datetime
goodDayNew = [[],[],[],[],[]]
for i in range(1,len(goodDay)):
    day = goodDay[i][0]
    day = day.replace('jan','1').replace('feb','2').replace('mar','3').replace('apr','4').replace('may','5').replace('jun','6').replace('jul','7').replace('aug','8').replace('sep','9').replace('oct','10').replace('nov','11').replace('dec','12')
    wholeTime = datetime.strptime(day,'%d %m %Y')
    month = str(wholeTime.strftime('%m'))
    year = str(wholeTime.strftime('%Y'))
    daytime = int(wholeTime.strftime('%d'))
    if year == '2017':
        if month == '10':
            if daytime >14:
                result = str(wholeTime.strftime('%y-%m-%d'))
                goodDayNew[0].append(result)
        if month == '11':
            result = str(wholeTime.strftime('%y-%m-%d'))
            goodDayNew[1].append(result)
        if month == '12':
            result = str(wholeTime.strftime('%y-%m-%d'))
            goodDayNew[2].append(result)
    if year == '2018':
        if month == '01':
            result = str(wholeTime.strftime('%y-%m-%d'))
            goodDayNew[3].append(result)
        if month == '02':
            result = str(wholeTime.strftime('%y-%m-%d'))
            goodDayNew[4].append(result)
```
