## case plan calculating score

```python

case_score = [0] * len(caseplan)
index = 0
#total score = 7
for row in caseplan.itertuples():
    sub = False
    mental = False
    physical = False
    if row[10] != 'Yes':#owe some money
        case_score[index] += 1
    if row[6] == 'No':#self-care
        case_score[index] += 1
    if row[17] == 'Yes' or row[18] == 'Yes': #substance use
        case_score[index] += 1
        sub = True
    if row[13] == 'Yes' or row[14] == 'Yes' or row[15] == 'Yes': #mental health
        case_score[index] += 1
        mental = True
    if row[16] == 'Yes' or row[7] == 'Yes' or row[11] == 'Yes':#physical health
        case_score[index] += 1
        physical = True
    if mental and sub and physical:# Tri-morbidity 
        case_score[index] += 1
    if row[11] == 'Yes' or row[7] == 'Yes' or row[16] == 'Yes':#Physical Health
        case_score[index] += 1
    index += 1
```

```python
caseplan['score'] = case_score
caseplan
```

```python
Intake_score = [0] * len(Intake)
index = 0
#total score = 5
for row in Intake.itertuples():
    if type(row[1]) == str:#History of homelessness
        if 'year' in row[1]:
            Intake_score[index] += 1
    if row[39] == 1 or row[40] == 1 or row[41] == 1 or row[42] == 1 or row[43] == 1 or row[44] == 1 or row[45] == 1 or\
         row[47] == 1 or row[48] == 1 or row[49] == 1 or row[50] == 1 or row[51] == 1 or row[52] == 1:
            #support self
            Intake_score[index] += 1
    if row[38] == 'Yes':#legal issues
        Intake_score[index] += 1
    if row[11] == 'Yes':#Chronic homelessness definition
        Intake_score[index] += 1
    if row[17] =='Yes':#physical health, do you have kids
        Intake_score[index] += 1
    index += 1
```

## New Intake score calculation

```python
import re
new_score = [0] * len(Intake_new)
chronic = [0]*len(Intake_new)
index = 0
# total socre = 5
for row in Intake_new.itertuples():
    if type(row[40]) == str:# How long since permanent stable housing?
        if 'year' in row[40]:
            homelessYear = True
        else:
            homelessYear = False
    if type(row[41]) == str:#Number of times homeless in last 3 years
        number = re.findall(r"\d+\.?\d*",row[41])
        if len(number) == 0:
            if 'once' in row[41] or 'twice' in row[41] or 'Once' in row[41]:
                homlessTimes = False
            else:
                homelessTime = True
        else:
            if int(number[0]) >= 3:
                homelessTime = True
            else:
                homelessTime = False
    else:
        homelessTime = False
    if homelessTime and homelessYear: #History of homelessness
        new_score[index] += 1
        chronic[index] = 1
    if type(row[25]) == str:
        if 'Yes' in row[25]: #Overnight in juvenile detention?
            new_score[index] += 1
    if row[27] == 1 or row[28] == 1 or row[29] == 1 or row[30] == 1 or row[31] == 1 or row[33] == 1 or row[34] == 1 or\
         row[35] == 1 or row[36] == 1 or row[37] == 1 or row[38] == 1 or row[39] == 1:#support self
            new_score[index] += 1 
    if row[58] == 1 or row[59] == 1 or row[60] == 1 or row[61] == 1:#Social Relationships
        new_score[index] += 1
    if row[63] == 1 or row[64] == 1: #Abuse/Trauma
        new_score[index] += 1
    index += 1
```

```python
Intake_new['Chronic homelessness definition'] = chronic
Intake_new['score'] = new_score
```

## combine different dictionary together and calculate total score

```python
class AddTest(object):
    def __init__(self,a):
        self.a=a
        assert isinstance(a,dict)
 
    def __str__(self):
        return self.a
 
    def __add__(self, other):
        for i in self.a.keys():
            if i != 'average':
                if other.a.get(i):
                    self.a[i] = self.a.get(i) +other.a.get(i)
                else:
                    self.a[i] = self.a.get(i) + other.a.get('average')
        for i in other.a.keys():
            if not self.a.get(i):
                self.a[i] = other.a.get(i) + self.a.get('average')
        self.a['average'] = self.a.get('average') + other.a.get('average')
        return self.a
    __repr__ = __str__
    
# a={"a":1,"b":2,"d":9,"average":4}
# b={"c":1,"a":2,"e":6,"average":3}
# c={"b":1,"f":2,"e":6,"average":3}
# d = AddTest(b)+AddTest(a)
# AddTest(d) + AddTest(c)
```

```python
sum_case = 0
for i in ID_score_case:
    sum_case += ID_score_case[i]
a_case = sum_case/len(ID_score_case)
ID_score_case['average'] = a_case
```

```python
sum_entrance = 0
for i in ID_score_entrance:
    sum_entrance += ID_score_entrance[i]
a_entrance = sum_entrance/len(ID_score_entrance)
ID_score_entrance['average'] = a_entrance
```

```python
sum_newIntake = 0
for i in ID_score_newIntake:
    sum_newIntake += ID_score_newIntake[i]
a_newIntake = sum_newIntake/len(ID_score_newIntake)
ID_score_newIntake['average'] = a_newIntake
```

```python
tmp = AddTest(ID_score_case) + AddTest(ID_score_entrance)
total_new = AddTest(tmp) + AddTest(ID_score_newIntake)
```
